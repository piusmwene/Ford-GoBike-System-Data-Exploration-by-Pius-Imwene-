```python
#making the copy of original bike_data dataset
bike_data_copy= bike_data.copy()
```

```python
#drop rows with missing values
bike_data_copy.dropna(inplace=True, axis=0)
```

```python
#check the sum of missing values in bike_data_copy
sum(bike_data_copy.isnull().sum())
```

```python
#Convert the start time, end time columns into datetime.
bike_data_copy['start_time'] = pd.to_datetime(bike_data_copy['start_time'])
bike_data_copy['end_time'] = pd.to_datetime(bike_data_copy['end_time'])
```

```python
#convert start_time and end_time into start_day and end_day
bike_data_copy['start_day'] = bike_data_copy['start_time'].dt.day_name()
bike_data_copy['end_day'] = bike_data_copy['end_time'].dt.day_name()
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
bike_data_copy['start_day'] = pd.Categorical(bike_data_copy['start_day'], categories=days, ordered=True)
bike_data_copy = bike_data_copy.sort_values('start_day')
```

```python
#Convert start_time and end_time into months.
bike_data_copy['start_month'] = bike_data_copy['start_time'].dt.month_name()
bike_data_copy['end_month'] = bike_data_copy['end_time'].dt.month_name()
```

```python
#convert start_station_id,end_station_id and bike_id of bike_data_copy into strings
bike_data_copy.start_station_id=bike_data_copy.start_station_id.astype(str)
bike_data_copy.end_station_id =bike_data_copy.end_station_id.astype(str)
bike_data_copy.bike_id =bike_data_copy.bike_id.astype(str)
```

```python
# convert seconds  into hours
bike_data_copy['duration_hours'] = bike_data_copy['duration_sec'] /3600
```

```python
#values count of the member_birth_year.
bike_data_copy['member_birth_year'].value_counts()
```

```python
#Replace 1878 with 1978 to be realistics.
bike_data_copy.replace(1878, 1978, inplace=True)
```

```python
# creating the column for age of the participants
bike_data_copy['participant_age']=(2019-bike_data_copy.member_birth_year)
```

```python
#check the information of dataset after preliminarly data wrangling.
bike_data_copy.info()
```

## Univariate Exploration

```python
#create a functions that will apply to all plot for xlabel, ylabel and title
def a_b_c(Alab, Blab, title):
    plt.title(title)
    plt.xlabel(Alab)
    plt.ylabel(Blab)
```

**The distribution of duration_sec**

```python
#histogram for the distribution of duration_sec
bins= np.arange(30, bike_data_copy['duration_sec'].max()+150, 100)
plt.figure(figsize=(15, 10))
plt.hist(data=bike_data_copy, x = 'duration_sec', bins=bins)
plt.xlim(0, 7000)
a_b_c('duration in seconds',' Frequency', 'The distribution of duration in seconds for the trip')
plt.show()
```

```python
# Since the graph has log tail, we transform using log transformation to understand data easily.
plt.figure(figsize=(20, 10))
bins = 10**np.arange(1.5,  4+0.1 , 0.1)
plt.hist(data = bike_data_copy , x = 'duration_sec' , bins = bins);
plt.xscale('log');
plt.xticks([100, 200, 400, 800, 1600, 3200, 7200],[100, 200, 400, 800, 1600, 3200, 7200])
a_b_c('duration in seconds',' Frequency', 'The distribution of duration in seconds for the trip');
```

In the first histogram of time (duration_sec), the disitribution of time (duration_sec) is positively skewed. It indicates that many participants spent lesser time while few particiants spent longer time. After, transforming durations_sec into log scale, the shape of the histogram became symmentric, hence, the distribution of time is normal with peak lying within 400 and 800.

**The distribution of participants_age**

```python
#histogram of participants age
bin1 = np.arange(15,130+10, 10 )
plt.figure(figsize=(12, 8))
plt.hist(data=bike_data_copy, x = 'participant_age', bins=bin1)
plt.xlim(10,100)
#add xlabel, ylabel and title
a_b_c('Age of Partcipants (years)', 'Frequency', 'The Distribution of Participants age')
plt.show()
```

```python
#histogram of participants age after using log scale
bin1 = 10**np.arange(1,4+0.07, 0.07 )
plt.figure(figsize=(10, 5))
plt.hist(data=bike_data_copy, x = 'participant_age', bins=bin1)
ticks=[10,25 , 40, 55, 70, 85,105]
labels = ['{}'.format(m) for m in ticks]
plt.xscale('log')
plt.xticks(ticks,labels)
a_b_c('log-participant age (years)', 'Frequency', 'The distribution of Participants log-age')
plt.show()
```

Based on the first histogram of participants_age, the figure has an assymetric shape. The distribution of age of the participants skewed to the right, thus, many participants are age between 15 to 45 years. However, on the second shape age of the particiapnts which has transformed by log scale, it has a symmentric/unimode shape with the peak between  25 and 40.

## Bivariate Exploration

**The relationship between gender and user_type**

```python
#clustered bar charts for user_type and gender
plt.figure(figsize=(12,6))
sb.countplot(data=bike_data_copy, x ='user_type', hue='member_gender')
a_b_c('Gender of Participants', 'Frequency', 'Clustered bar chart of user type and gender')
plt.show()
```

From the clustered bar chart, it indicate that there is a relationship between user type and gender. For example, males have a higher relationship with user type especially in subscriber than females.

**The relationship between duration time (sec) and the user type**

```python
# removing warning by importing libary warnings.
import warnings
warnings.filterwarnings('ignore')
#creating box plot for the relationship between duration_sec and user_type
color=sb.color_palette()[0]
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
sb.boxplot(data=bike_data_copy, x ='user_type', y='duration_sec', color=color)
a_b_c('user type', 'Trip duration (sec)', 'Box plot') 
plt.ylim(0, 5500)

plt.subplot(1,2,2)
sb.violinplot(data=bike_data_copy, x ='user_type', y='duration_sec', color=color, inner='quartile')
a_b_c('User type', 'Duration time (sec)', 'Violin Plot')
plt.ylim(0, 5500)
plt.show()
```

According to the violin and box plots, the customer spend more time to completed their trips compared to the subscriber.

**The relationship between duration time (sec) and bike_share_for_all_trip**

```python
#creating box plot for the relationship between duration time (sec) and bike_share_for_all_trip
color=sb.color_palette()[0]
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sb.violinplot(data=bike_data_copy, x ='bike_share_for_all_trip', y='duration_sec', color=color, inner='quartile')
a_b_c('bike share for all trips', 'Duration time (sec)', 'Violin Plot')
plt.ylim(0, 5500)

plt.subplot(1,2,2)
sb.boxplot(data=bike_data_copy, x ='bike_share_for_all_trip', y='duration_sec', color=color)
a_b_c('bike share for all trips', 'Duration time (sec)', 'Box plot') 
plt.ylim(0, 5500)
plt.show()
```

Based on the boxplot and violin plot, there is no difference for the user who share bike  and the users who dont share  for all trips.

**The relationship between duration time (sec) and end_day of trip**

```python
#violin and box plots for the relationship between duration_sec and end_day
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sb.violinplot(bike_data_copy['start_day'], bike_data_copy['duration_sec'], color=color, inner='quartile')
plt.xticks(rotation=15)
plt.ylim(0, 5500)
a_b_c('End day', 'Duration time (sec)', 'violin Plot')

plt.subplot(1,2,2)
sb.boxplot(bike_data_copy['end_day'], bike_data_copy['duration_sec'], color=color)
plt.xticks(rotation=15)
plt.ylim(0, 5500)
a_b_c('End day', 'Duration time (sec)', 'Box Plot')

plt.show()
```

From the violin and box plots above, Saturday and Sunday have more durations time at the end day compared to other days of the week.

**The relationship between participants' age and durations times(s) after transformed using log scale and inverse**

```python
# scatterplot of participants age and durations times after transformed using log scale and inverse.
plt.figure(figsize=(10,5))
df=bike_data_copy.sample(3000, replace=False)
def log(x, inverse=False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)
sb.regplot(df['participant_age'].apply(log), df['duration_sec'].apply(log),  scatter_kws={'alpha':1/4}, x_jitter=0.3)
a_b_c('Participant Age', 'Duration time (sec)', 'Scatter diagram of log scale of participants age vs. durations time (sec)')
plt.show()
```

From the scatterplot of log of participats age and log of durations time (sec), it indicate that there is moderate relatioship.

## Multivariate Exploration

**The relationship between participants age and duration time (sec) faceted by user type**

```python
# getting a sample of 1000 from bike_data_copy for scatterplot.
sample_bike_data=bike_data_copy.sample(n=1000, replace = False)
#scatter plot of participants age vs.duration time faceted by user type
graph = sb.FacetGrid(data = sample_bike_data, col = 'user_type', col_wrap=3, size = 5, sharey=False)
graph.map(plt.scatter, 'participant_age', 'duration_sec', alpha =0.7)
graph.set(ylim=(0, 5500))
graph.set(ylabel='Duration time (sec)')
graph.set(xlabel='Age of Participants')
plt.show();
```

The two scatterplots indicates a postive relationship between age of participants and duration time in second for trip. Howerver, there is a stronger relationship between age of participants and customer  compared to subscriber and participants' age.

**The relationship between participants age and durations time (sec) influenced by user type**

```python
#create age intervals qith 5 years intervals
plt.figure(figsize=(15,10))
bike_data_copy['participant_age_int']=((bike_data_copy['participant_age']//5)*5).astype(int)
sb.stripplot(data=bike_data_copy.query('duration_sec<duration_sec.quantile(0.99) and participant_age<80'),
             x='participant_age_int', y='duration_sec', hue='user_type', size=2, jitter=0.35, dodge=True)
a_b_c('Participant Age', 'Duration time (sec)', 'The relationship between duration time (sec) vs. Partcipants Age by user type')
plt.show();
```

In scatterplot above shows the relationship between age of participants and the durations time (sec) that influenced by user type. Based on the scattter diagram, there is a strong relationship between participants' age and customer.

**The relationship between participants age and durations time (sec) influenced by member gender**

```python
#create age intervals qith 5 years intervals
plt.figure(figsize=(15,10))
bike_data_copy['participant_age_int']=((bike_data_copy['participant_age']//5)*5).astype(int)
sb.stripplot(data=bike_data_copy.query('duration_sec<duration_sec.quantile(0.99) and participant_age<80'),
             x='participant_age_int', y='duration_sec', hue='member_gender', size=2, jitter=0.35, dodge=True)
a_b_c('Participant Age', 'Duration time (sec)', 'The relationship between duration time (sec) vs. Partcipants Age by gender')
plt.show();
```

The diagram above shows the relationship between participants age and duration time (sec) by gender. From, the diagram, there was strong relationship between participants' age and duration time in male than in female.
